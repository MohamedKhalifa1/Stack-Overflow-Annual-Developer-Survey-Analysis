```python
RANDOM_STATE = 42
ROLE_COLS      = ['DevType']
TECH_COLS      = ['LanguageHaveWorkedWith',
                  'DatabaseHaveWorkedWith',
                  'PlatformHaveWorkedWith',
                  'WebframeHaveWorkedWith',
                  'MiscTechHaveWorkedWith',
                  'ToolsTechHaveWorkedWith',
                  'NEWCollabToolsHaveWorkedWith']
MLFLOW_TRACKING_URL = '/content/Stack-Overflow-Annual-Developer-Survey-Analysis/models/mlruns'
DATA_PATH = '/content/Stack-Overflow-Annual-Developer-Survey-Analysis/data/preprocessed/04_cleaned_data.pkl.gz'
MLFLOW_EXPEREMNET_NAME = 'skills_job_stackoverflow'
LOG_PATH = '/content/Stack-Overflow-Annual-Developer-Survey-Analysis/models/temp'
LOG_DATA_PKL    =  "data.pkl"
LOG_MODEL_PKL   =  "model.pkl"
LOG_METRICS     =  "metrics.pkl"
```

```python
import gzip
import os
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import seaborn as sns
from sklearn.model_selection import train_test_split
from mlflow.tracking import MlflowClient
import mlflow
import mlflow.sklearn
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.multioutput  import MultiOutputClassifier
from sklearn.metrics import auc, accuracy_score, confusion_matrix, f1_score, precision_score, recall_score
```

```python
with gzip.open(DATA_PATH,'rb') as f:
    df = pickle.load(f)
```

```python
df.head()
```

```python
df = df.loc[:, (df != '').any(axis=0)]
df.head()
```

```python
def calculate_quality(ground_truth, prediction , metric_function , sort_values = False):
  quality_scores = {}

  for col in prediction.columns :
    role_pred = prediction[col].copy()
    role_truth = ground_truth[col].copy()
    quality_scores[col] = round(metric_function(role_truth, role_pred)*100, 2)

  if sort_values :
    quality_scores = quality_scores.sort()

  return quality_scores
```

```python
Path(Log_PATH).mkdir(parents=True, exist_ok=True)
Path(MLFLOW_TRACKING_URL).mkdir(parents=True, exist_ok=True)
```

```python
role_df = df['DevType'].copy()
role_df.sum()
```

```python
from os import replace
sample_per_class = 1100
resample_roles = []

for role_col in role_df.columns:
  sub_df = role_df.loc[role_df[role_col] == 1].copy()

  if len(sub_df) > sample_per_class :
    sub_df = sub_df.sample(sample_per_class , replace = True , random_state = RANDOM_STATE)
    resample_roles.append(sub_df)
  if len(sub_df) < sample_per_class :
    sub_df = sub_df.sample(sample_per_class , replace = True , random_state = RANDOM_STATE)
  resample_roles.append(sub_df)
```

```python
role_df = pd.concat(resample_roles)
df = df.loc[role_df.index].copy()
```

```python
df.head()
```

```python
role_df.sum()
```

```python
X = df.drop(columns = ['DevType'], axis = 1)
y = df['DevType']
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.2 , random_state = RANDOM_STATE)
```

```python
X_train.shape , X_test.shape , y_train.shape , y_test.shape
```

```python
mlflow.set_tracking_uri(MLFLOW_TRACKING_URL)
client = MlflowClient()
mlflow.get_experiment_by_name(MLFLOW_EXPEREMNET_NAME)
exp = client.get_experiment_by_name(MLFLOW_EXPEREMNET_NAME)
if exp is None:
    exp_id = mlflow.create_experiment(MLFLOW_EXPEREMNET_NAME)
else:
    exp_id = exp.experiment_id
```

```python
clf = make_pipeline(StandardScaler(),MultiOutputClassifier(LogisticRegression(random_state=RANDOM_STATE)) )
```

```python
clf.fit(X_train.values , y_train.values)
```

```python
y_pred = clf.predict(X_train)
```

```python
predictions = pd.DataFrame(y_pred , columns = y_train.columns)
```

```python
predictions
```

```python
prediction = pd.DataFrame(y_pred, columns=y_train.columns)
train_scores = {score.__name__: calculate_quality(y_train, prediction, score) for score in [accuracy_score, precision_score, recall_score, f1_score]}
train_scores_df = pd.DataFrame(train_scores)
train_scores_df
```

```python
prediction = pd.DataFrame(clf.predict(X_test), columns=y_test.columns)
test_scores = {score.__name__: calculate_quality(y_test, prediction, score) for score in [accuracy_score, precision_score, recall_score, f1_score]}
test_scores_df = pd.DataFrame(test_scores)
test_scores_df
```

```python
mean_test_scores = test_scores_df.mean()
print(mean_test_scores)
```

```python
data_details = {
    'data_path':DATA_PATH ,
    'training_indices' : X_train.index.tolist(),
    'testing_indices' : X_test.index.tolist(),
    'feature_name' :X_train.columns.droplevel(0).tolist(),
    'target_name' :y_train.columns.tolist(),
}
with open(os.path.join(LOG_PATH, LOG_DATA_PKL), 'wb') as f:
    pickle.dump(data_details, f)
```

```python
model = {"model_description": "Baseline model: Logistic Regression ",
         "model_details": str(clf),
         "model_object": clf}
with open(os.path.join(LOG_PATH, LOG_MODEL_PKL), 'wb') as f:
    pickle.dump(model, f)
```

```python
classes_metrics = {"train_scores": train_scores_df,
                   "test_scores":  test_scores_df}

with open(os.path.join(LOG_PATH, LOG_METRICS), "wb") as f:
    pickle.dump(classes_metrics, f)
```

```python
if exp_id:
    with mlflow.start_run(experiment_id=exp_id, run_name=model["model_description"]):
        mlflow.log_artifacts(LOG_PATH)
        for metric, score in mean_test_scores.items():
            mlflow.log_metric(metric, score)
else:
    print(f"Failed to fetch or create experiment '{MLFLOW_EXPEREMNET_NAME}'")
```
