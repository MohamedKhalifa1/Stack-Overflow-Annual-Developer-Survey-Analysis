```python
import pandas as pd
import mlflow
from mlflow.tracking import MlflowClient
import pickle
import os
import yaml
```

```python
model_path = os.path.join(ARTIFACT_PATH, LOG_MODEL_PKL)
with open(model_path, 'rb') as f:
    model = pickle.load(f)
```

```python
model.keys()
```

```python
data_path = os.path.join(ARTIFACT_PATH, LOG_DATA_PKL)
with open(data_path, 'rb') as f:
    data = pickle.load(f)
```

```python
data.keys()
```

```python
features_names = pd.Series(data["feature_name"])
targets_names  = pd.Series(data['target_name'])
classifier = model['model_object']
```

```python
with open(CLUSTERS_YAML_PATH, 'r') as f:
    clusters_config = yaml.safe_load(f)
clusters_config
```

```python
reformat_cluster = [(cluster_name, cluster_skill)
                   for cluster_name, cluster_skills in clusters_config.items()
                   for cluster_skill in cluster_skills]
reformat_cluster
```

```python
clusters_df = pd.DataFrame(reformat_cluster, columns=['cluster_name', 'skills'])
clusters_df
```

```python
sample_skills = [
    'Scikit-Learn', 'Hugging Face Transformers', 'Python' ,'Scala' ,'Hadoop',
    'Jupyter Notebook/JupyterLab','PyCharm' ,'RStudio' ,'Django']
```

```python
pd.Series(sample_skills).isin(features_names)
```

```python
sample_cluster = clusters_df.copy()
sample_cluster['is_skill'] = sample_cluster['skills'].isin(sample_skills)
sample_cluster
```

```python
sample_cluster.is_skill.value_counts()
```

```python
cluster_features = sample_cluster.groupby("cluster_name")["is_skill"].sum()
cluster_features
```

```python
~features_names.isin(cluster_features.index)
```

```python
skills_names = features_names[~features_names.isin(cluster_features.index)]
```

```python
ohe_skills = pd.Series(skills_names.isin(sample_skills).astype(int).tolist(),
                       index=skills_names)
ohe_skills
```

```python
features = pd.concat([ohe_skills,
                      cluster_features])
```

```python
features.loc['Scala']
```

```python
features = features.loc[features_names]
features
```

```python
predictions = classifier.predict_proba([features.values])
predictions
```

```python
positive_probs = [prob[0][1] for prob in predictions]
pd.Series(positive_probs,
          index=targets_names).sort_values(ascending=False)
```
